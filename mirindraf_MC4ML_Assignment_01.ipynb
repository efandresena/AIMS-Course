### Q2. Gaussian sums.


#### (a). Sum of two variables:

We consider the sum of two IID Gaussian random variables as Z = X + Y.
Using the fourier transform function of Z yields :

  $$
  \begin{align*}
  {\varphi}_{Z} (k) &=\int_{\infty} e^{ikz}p(z) dz =\mathbb{E}( e^{ikz}) \\
  &= \mathbb{E}( e^{ik(x+y)})\\
  &= \mathbb{E}( e^{ikx})\mathbb{E}( e^{iky}) \text{  Since they are independant}\\
  &= {\varphi}_{X} (k) {\varphi}_{Y} (k)
  \end{align*}
  $$


Using the result in Q1.c.1,  assuming both have the same mean and variance  

  $$
  \begin{align*}
  {\varphi}_{Z} (k) &= e^{- \frac{1}{2}k^2\sigma^2 + ik\mu}  e^{- \frac{1}{2}k^2\sigma^2 + ik\mu} \\
  \end{align*}
  $$

  Taking the inverse fourier transform to get back to $p(z)$

  $$
  \begin{align*}
  p(z) &= ∫_{∞} e^{-izk}{\varphi}_{Z} (k) dk\\
  &= ∫_{∞} e^{-izk} e^{-k^2\sigma^2 + 2ik\mu} dk\\
  &= ∫_{∞}  e^{-k^2\sigma^2 + ik(2\mu - z) } dk
  \end{align*}
  $$

We will use the method completing the square :  

$$
\begin{align*}
 -k^2\sigma^2 + ik(2\mu - z)
 &= -\sigma^2 \left(k^2 - i k \frac{(2\mu - z)}{\sigma^2}\right) \\
 &= -\sigma^2 \left[\left(k - i \frac{(2\mu - z)}{2\sigma^2}\right)^2 - \left(\frac{(2\mu - z)}{2\sigma^2}\right)^2\right] \\
 &= -\sigma^2 \left(k - i \frac{(2\mu - z)}{2\sigma^2}\right)^2 + \frac{(2\mu - z)^2}{4\sigma^2}.
\end{align*}
$$

Using the Gaussian integral (and allowing a complex shift for \(k\)):

$$
\begin{align*}
p(z)
&= \int_{-\infty}^{\infty} e^{-k^2\sigma^2 + ik(2\mu - z)} \, dk \\
&= e^{-\frac{(z - 2\mu)^2}{4\sigma^2}}
    \int_{-\infty}^{\infty} e^{-\sigma^2 \left(k - i \frac{(2\mu - z)}{2\sigma^2}\right)^2} \, dk \\
&= \frac{1}{\sqrt{4\pi\sigma^2}}
   \exp\!\left[-\frac{(z - 2\mu)^2}{4\sigma^2}\right].
\end{align*}
$$

Thus the resulting distribution is $ Z \sim \mathcal{N}(\mu + \mu,\, \sigma^2 + \sigma^2)  $ $ \sim \mathcal{N}(2\mu,\, 2\sigma^2) $.

Q.E.D



#### (b). Generalisation to n random variables:


We will proceed by induction, as we have already proven that sum of two variables X and Y can be considered as the transformation $X \to Z = X + Y$, given they are IID. Hence for 3 variables we can consider the transformation $X \to Z = X + Y \to G = Z + W$.  With $ Z \sim \mathcal{N}(2\mu,\, 2\sigma^2).$ and $  W \sim \mathcal{N}(\mu,\, \sigma^2) $.

Wich will make $  G \sim \mathcal{N}(3\mu,\, 3\sigma^2) $

By extension, for $n$ IID Gaussian random variables $X_1, X_2, \dots, X_n \sim \mathcal{N}(\mu, \sigma^2)$, their sum

$$
S_n = \sum_{i=1}^{n} X_i
$$

follows the distribution

$$
S_n \sim \mathcal{N}(n\mu,\, n\sigma^2).
$$

```python
# Implementation
def myhistogram(v:iter, a, b, n:int):
  dx = (b-a)/n
  counts = []
  bins =[a]
  position = a
  for _ in range(n):
    count = sum(1 for e in v if position<=e<=(position+dx))
    counts.append(count)
    position += dx
    bins.append(position)
  return (np.array(counts, dtype=int), np.array(bins, dtype=float))
```

### Q5. Non-uniform variates.

#### (a). Random choice in a list with each element having Pr= pi

The CDF is defined as :
$$
\begin{align*}
F(x) = \begin{cases}
  p_n, & \text{if } x=n \\
\end{cases}
\end{align*}
$$

Taking the inverse with u sampled from uniform U[0,1]
$$
\begin{align*}
F^{-1}(u) = \begin{cases}
  n, & \text{if } \sum_{0}^{n-1} p_n < u \le \sum_{0}^{n} p_n \\
\end{cases}
\end{align*}
$$

```python
def random_choice(pr, n_samples=1):
  samples = []
  for _ in range(n_samples):
    u = np.random.random()

    for i in range(1,len(pr)+1):
      sum_left = sum(pr[:i-1])
      sum_right = sum(pr[:i])
      if sum_left<u<=sum_right:
        x = i
    samples.append(x)
  return np.array(samples, dtype=int)
```

```python
n = 50
n_samples = 1000
random_values = np.random.rand(n)
pr = random_values / random_values.sum()
generated = random_choice(pr, n_samples)

# Compute histogram of generated data
counts, bin_edges = np.histogram(generated, bins=n, density=True)
a, b = min(bin_edges), max(bin_edges)
dx = (b-a)/n
bin_centers = bin_edges[:-1] - (dx/2)


plt.figure(figsize=(8,6))

# Plot generated distrubition
plt.bar(bin_centers, counts, width=dx, color='skyblue', edgecolor='k', alpha=0.7, label='Generated')

# Plot theoretical distribution
plt.bar(range(n), pr, color='orange', alpha=0.5, label='Theoretical')

plt.xlabel("Value")
plt.ylabel("Probability")
plt.title("Random Choice Distribution")
plt.legend()
plt.show()
```

#### (b). Uniform over [-1,1]


The CDF is defined as :

$$
\begin{align*}
F(x) = ∫_{-1}^{x} \frac{1}{2} dk = \frac{1}{2} (x +1)
\end{align*}
$$

Taking the inverse with u sampled from uniform U[0,1]
$$
\begin{align*}
F^{-1}(u) = 2u-1

```python
def uniform_1(n_samples=1):
  samples = []
  for _ in range(n_samples):
    u = np.random.random()
    x = 2*u - 1
    samples.append(x)
  return np.array(samples, dtype=float)
```

```python
def compare_hist(implemented_samples, theoretical_samples, n_bins=100, dist_title='_'):
    # Determine common range for bins
    a = min(implemented_samples.min(), theoretical_samples.min())
    b = max(implemented_samples.max(), theoretical_samples.max())

    # Compute histogram of implemented/generated samples
    counts_imp, bin_edges = np.histogram(implemented_samples, bins=n_bins, range=(a,b), density=True)
    counts_th, _ = np.histogram(theoretical_samples, bins=bin_edges, density=True)

    # Compute bin centers
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    dx = bin_edges[1] - bin_edges[0]

    # Plot
    plt.figure(figsize=(8,6))
    plt.bar(bin_centers, counts_imp, width=dx, color='skyblue', edgecolor='k', alpha=0.7, label='Generated')
    plt.bar(bin_centers, counts_th, width=dx, color='red', alpha=0.2, label='Theoretical')

    plt.xlabel("Value")
    plt.ylabel("Probability Density")
    plt.title(f"{dist_title} Distribution Comparison")
    plt.legend()
    plt.show()



n_samples = 10000
uniform_implemented = uniform_1(n_samples)
th_uniform = np.random.uniform(-1,1,n_samples)
compare_hist(uniform_implemented, th_uniform, n_bins=50, dist_title='Uniform')
```

#### (c). Exponential with parameter $\lambda$

The CDF is defined as :

$$
F_X(x) =
\begin{cases}
    0 & \text{if } x < 0 \\
    1 - e^{-\lambda x} & \text{if } x \ge 0
\end{cases}
$$

Taking the inverse with u sampled from uniform U[0,1]
$$
\begin{align*}
F^{-1}(u) = \frac{-log(u)}{λ}
\end{align*}
$$

```python
def exponential(Lambda, n_samples=1):
  samples = []
  for _ in range(n_samples):
    u = np.random.random()
    x = -np.log(u)/Lambda
    samples.append(x)
  return np.array(samples, dtype=float)
```

```python
Lambda = 1.5
expo_implemented = exponential(Lambda, n_samples)
th_expo = np.random.exponential(Lambda, n_samples)
compare_hist(expo_implemented, th_expo, n_bins=50, dist_title='Exponential')
```

```python
def generate_r(n_samples=1):
  u1 = np.random.uniform(0,1,n_samples)
  r = np.sqrt(-2*np.log(u1))
  return r

def generate_theta(n_samples=1):
  u2 = np.random.uniform(0,1,n_samples)
  theta = 2*np.pi*u2
  return theta

```



*   Generate a Gaussian distrubition sampled as $X ∼(\mu, \sigma^2)$





```python
def generate_gaussian( n_samples=1, mu=0, var=1,):
  r = generate_r(n_samples)
  theta = generate_theta(n_samples)
  x = np.sqrt(var)*r*np.cos(theta) + mu
  return x
```


### Q7. Monte Carlo estimation of $\pi$.

#### (a). Algorithm
The probability of the random point being in the circle is related to the ration of the number of element in the circle and versus in the whole square. $\Pr(in\_circle) = \frac{n_circle}{n_square} = \frac{π}{4}$

```python

def approximate_pi():

  n_square = 100000
  n_circle = 0
  x = np.random.uniform(-1,1, n_square)
  y = np.random.uniform(-1,1, n_square)
  C = np.pow(x, 2.0) + np.pow(y, 2.0)
  n_circle = np.sum(C <= 1.0)
  pi = (4*n_circle)/n_square
  return pi

pi = approximate_pi()
print('Approximated Value of PI: ',pi)
```

#### (b). Statistical test

```python
# Point estimator
L = 300
for l in range(0,L+1, L//3):
  pi_samples = np.array([approximate_pi() for _ in range(l)])
  mean = pi_samples.mean()
  print('L: ', l)
  print('Mean Estimate: ', mean)
```

```python
# Mean estimator

Lmax = 10**3
np.random.seed(10)

def mean_estimator(approximate_RV, Lmax):
  # Empty lists to receive results
  mean_est_list = np.zeros(Lmax)
  mean_est_err_list = np.zeros(Lmax)

  # Initial values of sums
  s = 0.0
  v = 0.0
  for i in range(Lmax):
    RV = approximate_RV()
    s += RV # Accumulate the RVs in sum to get mean estimator
    v += RV**2 # Accumulate the RVs^2 to get the error bar
    mean_est = s/(i+1.0) # Mean estimator
    sec_mom_est = v/(i+1.0) # Second moment estimator
    mean_est_err = np.sqrt(sec_mom_est-mean_est**2) / np.sqrt(i+1.0) # Estimator standard deviation
    # Put results in lists
    mean_est_list[i] = mean_est
    mean_est_err_list[i] = mean_est_err

  return mean_est_list, mean_est_err_list

mean_est_list, mean_est_err_list = mean_estimator(approximate_pi, Lmax)
```

```python
def plot_convergence(mean_est_list, mean_est_err_list, true_value=None, plot_label="Estimator"):
    """
    Plot Monte Carlo estimator convergence with error bars.

    Parameters
    ----------
    mean_est_list : array-like
        Sequence of mean estimates (running means).
    mean_est_err_list : array-like
        Sequence of standard errors corresponding to the mean estimates.
    true_value : float, optional
        Theoretical or expected value to plot as a reference line.
    plot_label : str, optional
        Label for the estimator in the plot.
    """
    Lmax = len(mean_est_list)

    plt.figure(figsize=(8, 5))
    plt.style.use('seaborn-v0_8-darkgrid')

    # Plot estimates with error bars
    plt.errorbar(
        range(Lmax),
        mean_est_list,
        yerr=mean_est_err_list,
        fmt='o',
        markersize=4,
        capsize=3,
        elinewidth=1,
        errorevery=max(1, Lmax // 100),  # adjust error bar density automatically
        color='tab:blue',
        ecolor='lightblue',
        label=plot_label
    )

    # Plot true (theoretical) value if provided
    if true_value is not None:
        plt.plot(
            range(Lmax),
            np.full(Lmax, true_value),
            'k--',
            linewidth=2,
            label='Theoretical value'
        )

    plt.xlabel('Number of simulations (L)')
    plt.ylabel('Estimated value')
    plt.title(f'Monte Carlo Convergence of {plot_label}')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
plot_convergence(mean_est_list, mean_est_err_list, true_value=np.pi, plot_label="PI")
```

### Q8. Two-state Markov chain.

#### (a). Algorithm Implementation


[Ref Algorithm : Simulating Markov chains 2007 by Karl Sigman University of Colombia](http://www.columbia.edu/~ks20/4703-Sigman/4703-07-Notes-MC.pdf)

```python
def bernouilli(p):
  u = np.random.uniform()
  if u<p:
    return 1
  else:
    return 0

def markov_chain(alpha, beta, n_path = 4):
  #alpha : 0->1
  #beta : 1->0
  path = [0]
  n = 1
  # generate :
  x0 = bernouilli(alpha)
  path.append(x0)

  while n<n_path-1:
    xi = path[-1]
    n+=1
    if xi == 0:
      next_state = bernouilli(alpha) # if we are on 0 we decide according to alpha wether going to 1 or stay at 0
    else:
      next_state = 1 - bernouilli(beta) # if we are on 1 we decide according to beta wether going to 0 or stay at 1
                                        # however bernouilli(beta) return 1 mean go to 0 so we need the inverse
    path.append(next_state)
  return path
```

#### (b).  Simulation

#### (c). Statistical Estimation and Verification

```python
# Verification of the long term occupations on one run:
n_path = 1e3
path = markov_chain(alpha, beta, n_path)
long_term_0 = beta/(alpha + beta)
long_term_1 = alpha/(alpha + beta)

# Estimation
count_0 = sum(1 for e in path if  (e==0))
long_term_0_est = count_0/n_path
long_term_1_est = 1 - long_term_0_est
print(f'True Values of long term occupations of (0,1) : {long_term_0, long_term_1}')
print(f"{'-'*60}")
print(f'Estimated Values of long term occupations of (0,1) : {long_term_0_est, long_term_1_est}')
```

```python
def approximate_lte(alpha=0.755, beta=0.123, n_path=int(1e3), param_0=False): # Need to tweak for each call (set to tru for param_1)
    """
    Estimate the long-term probability of being in state 1
    in a two-state Markov chain with parameters alpha, beta.
    """
    path = markov_chain(alpha, beta, n_path)
    count_0 = sum(e == 0 for e in path)
    long_term_0_est = count_0 / n_path
    long_term_1_est = 1 - long_term_0_est
    if param_0:
      return long_term_0_est
    else :
      return long_term_1_est
```

#### (a). Generate a connected graph 10 states

```python
import networkx as nx
import matplotlib.pyplot as plt

# Parameters
n_states = 10
p_connection = 0.4  # probability of edge creation

# Generate a connected random graph
G = nx.erdos_renyi_graph(n_states, p_connection)

# If it's not connected, keep generating until it is
while not nx.is_connected(G):
    G = nx.erdos_renyi_graph(n_states, p_connection)

# Draw the graph
plt.figure(figsize=(6, 6))
pos = nx.spring_layout(G, seed=42)
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color="skyblue",
    node_size=800,
    font_weight="bold",
    edge_color="gray"
)
plt.title("Connected Random Graph with 10 States")
plt.show()
```

#### (b). Program that simulates a trajectory using uniform distrubition

```python
def random_walk(G, start_node=None, n_steps=100):
    """
    Simulate a random walk on a (connected) graph G.

    Parameters
    ----------
    G : networkx.Graph
        The graph to walk on.
    start_node : int or None
        Starting node. If None, choose randomly.
    n_steps : int
        Number of steps in the walk.

    Returns
    -------
    list
        Sequence of visited nodes.
    """
    if start_node is None:
        start_node = np.random.choice(G.nodes())

    path = [start_node]

    current = start_node
    for _ in range(n_steps):
        neighbors = list(G.neighbors(current))
        current = np.random.choice(neighbors)
        path.append(current)

    return path
```

#### (c). Trajectory plot steps = 1000

```python
walk = random_walk(G, n_steps=1000)
print("Random walk path:", walk)
```

```python
plt.figure(figsize=(10,5))
plt.plot(np.arange(len(walk)), walk, color='blue', label='Random walk')
plt.xlabel('Step')
plt.ylabel('Node')
plt.xscale('log')
plt.title('Random Walk Trajectory')
plt.show()
```

#### (d). Proportionality between fration of visit and degree ki of a node

```python
# Get the degree of each nodes
degrees = G.degree()
degree = np.array([d for _,d in degrees], dtype=float)
degree
```

```python
# Get the number of occurence of each nodes
_, counts = np.unique(walk, return_counts=True)
```

```python
# Comput the fraction time of visit on node i
fraction = counts/len(counts)
fraction
```

```python
# Fit linear model: fraction ≈ k * degree
coeffs = np.polyfit(degree, fraction, 1)
slope, intercept = coeffs
print(f"Fitted line: fraction = {slope:.4f} * degree + {intercept:.4f}")
```

```python
# Generate line for plotting
degree_fit = np.linspace(min(degree), max(degree), 100)
fraction_fitted = np.polyval(coeffs, degree_fit)

# Plot
plt.figure(figsize=(8,6))
plt.scatter(degree, fraction, color='blue', label='Random walk data')
plt.plot(degree_fit, fraction_fitted, color='red', linewidth=2, label=f'Linear fit: y={slope:.2f}x+{intercept:.2f}')
plt.xlabel('Degree ')
plt.ylabel('Fraction of visit')
plt.title('Proportionality between degree and fraction')
plt.legend()
plt.tight_layout()
plt.show()
```
