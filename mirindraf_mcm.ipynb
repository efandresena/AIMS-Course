## Q1. Common random variable:  

* **a. Bernoulli (p):**  

  1. *Mean:*

  Computed using :
  $$
  \begin{align*}
  \mu &= \mathbb{E}[X] = \sum_{x \in \Omega} x \cdot \Pr(X = x) \\
      &= 0 \times (1 - p) + 1 \times p \\
  \mu &=p
  \end{align*}
  $$


  2. *Variance:*
  
  Given by:  
  $$
  \begin{align*}
  \sigma^2 &= \mathbb{E}[(X - \mu)^2] \\
  & \text{ or } \\
  &= \mathbb{E}[X^2] - (\mathbb{E}[X])^2 \\
  &= \sum_{x \in \Omega} x^2 \cdot \Pr(X = x) - p^2\\
  &= p - p^2\\
  \sigma^2&=p(1 - p)
  \end{align*}
  $$  

  3. *Function Characteristic:*
  
  Given by the fourier transform of the PMF

  $$
  f(x) =
  \begin{cases}
  p, & x = 1 \\
  1 - p, & x = 0
  \end{cases}
  $$
  
  $$
  \begin{align*}
  {\varphi}_X (k) &= ∑_{x} e^{ikx} Pr(X=x) \\
  &= (1-p) + pe^{ikt}
  \end{align*}
  $$

* **b. Binomial:**  
$$
 f(x) = \binom{n}{x} p^x (1 - p)^{n - x}
$$
  1. *Mean:*
$$
\begin{align*}
\mu &=\sum_{x=0}^{n} x \cdot \Pr(X = x) \\
&=\sum_{x=0} x \cdot \binom{n}{x} p^x (1 - p)^{n - x} \\
&=\sum_{x } x \cdot \frac{n!}{x!(n - x)!} p^x (1 - p)^{n - x} \\
&=\sum_{x }   \frac{n (n-1!}{(x-1)!(n - x)!} p^x (1 - p)^{n - x} \\
\text{ by noticing the binomial  } &\text{ combination term for n-1 and x -1  and also by factoring out n and set (x-1)-> k  we get }\\
&= n \sum_{k=1}^{n} \binom{n-1}{k} p^{k+1} (1 - p)^{n - k - 1} \\
&= np \sum_{k} \binom{n-1}{k} p^{k} (1 - p)^{n - k - 1}\\
\text{setting the dummy variable n-1->n}\\
&= np \sum_{k=0}^{n} \binom{n}{k} p^{k} (1 - p)^{n - k}\\
&= np \sum_{k=0}^{n} (p+(1-p))^n\\
&=np
\end{align*}
$$



  2. *Function Characteristic:*
  

  $$
  \begin{align*}
  {\varphi}_X (k) &= ∑_{x} e^{ikx} f(x) \\
  &= ∑_{x}  \binom{n}{x}  (1 - p)^{n - x} (pe^{ik})^x \\
  &= ((1-p) + pe^{ik})^n
  \end{align*}
  $$

    3. *Variance:*
  
  Using the function characteristic, we have by definition
  $$
  \begin{align*}
  \sigma^2 &= - φ^{"}_{X}{(0)} - {{φ^{'}_{X}}^2{(0)}}\\
  \text{ with } \\
  φ^{'}_{X}{(0)}&= inpe^{ik}[(1-p+e^{ik}]^{n-1} = inp\\
  φ^{"}_{X}{(0)}&=np+np(n-1)p \\
  \text{ hence                   }\\
  \sigma^2&=np(1 - p)
  \end{align*}
  $$

* **c. Gaussian:**  
$$
 f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \, e^{ -\frac{(x - \mu)^2}{2\sigma^2} }
$$

1. *Function Characteristic:*
  

  $$
  \begin{align*}
  {\varphi}_X (k) &= \int_{∞}dx e^{ikx} \frac{1}{\sqrt{2\pi\sigma^2}} \, e^{ -\frac{(x - \mu)^2}{2\sigma^2} } \\
  \end{align*}
  $$

  Using the completing square method to factorize the argument of the exponential,

  $$
  \begin{align*}
 ikx -\frac{(x - \mu)^2}{2\sigma^2} &= - \frac{1}{2\sigma^2} [(x - \mu)^2- 2ikx\sigma^2]\\
&= - \frac{1}{2\sigma^2} [(x - \mu)^2- ikx(2\sigma^2)]\\
&= - \frac{1}{2\sigma^2} [x^2 + \mu^2- 2x(\mu + ik \sigma^2)]\\
&= -\frac{1}{2\sigma^2}\left[(x - (\mu + ik\sigma^2))^2 - (\mu + ik\sigma^2)^2 + \mu^2\right] \\
&= -\frac{[x - (\mu + ik\sigma^2)]^2}{2\sigma^2} + ik\mu - \frac{1}{2}k^2\sigma^2
  \end{align*}
  $$

  Thus,
  $$
  \begin{align*}
  {\varphi}_X (k) &= e^{- \frac{1}{2}k^2\sigma^2 + ik\mu} \underbrace{\frac{1}{\sqrt{2\pi\sigma^2}} \int_{∞} dx \, e^{ -\frac{[x - (\mu + ik\sigma^2)]^2}{2\sigma^2} } }_{1}\\
  \end{align*}
  $$
Where we have assumed that in the last term the gauss integral holds even if the shift of the mean are complex. [1](https://en.wikipedia.org/wiki/Cauchy's_integral_theorem#:~:text=In%20mathematics%2C%20the%20Cauchy%20integral%20theorem%20in,in%20%CE%A9%2C%20that%20contour%20integral%20is%20zero.)

  2. *Mean:*
$$
\begin{align*}
 \mathbb{E}[X] &= \frac{1}{i}{φ^{'}_{X}}(0) \\
 &= \frac{1}{i}(iu-\sigma^2t)e^{iut-\frac{1}{2}\sigma^2t^2 |_{t=0}}\\
 &=u
\end{align*}
$$



  

  3. *Variance:*
  $$
  \begin{align*}
  - φ^{"}_{X}{(0)} &= -[(-\sigma^2)e^{\phi(0)} + (iu-\sigma^2t)^2e^{\phi(0)}]\\
  \mathbb{Var}{X} &= +\sigma^2 + \mu^2 - \mathbb{E}[X]^{2}\\
  \mathbb{Var}{X} &= \sigma^2
  \end{align*}
  $$
